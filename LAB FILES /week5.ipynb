### **Activity 5: Unique Hobbies Set** 


Two friends share their hobbies:
alex_hobbies = {"reading", "gaming", "cooking", "hiking"}
sam_hobbies = {"gaming", "hiking", "photography", "music"}

Find and print:
- Hobbies both enjoy (intersection)
- Hobbies only Alex enjoys
- All unique hobbies combined

---

`````python
alex_hobbies = {"reading", "gaming", "cooking", "hiking"} 
sam_hobbies = {"gaming", "hiking", "photography", "music"}

both= alex_hobbies.intersection(sam_hobbies) 
print("Hobbies that both alex and sam enjoy:", both) 

alex= alex_hobbies.difference(sam_hobbies) 
print("Hobbies only alex enjoys:", alex) 

 

all= alex_hobbies.union(sam_hobbies) 
print("all unique hobbies combined:", all) 
`````


### **Activity 6: Simple Grade Book Dictionary** 

Create a dictionary for 3 students and their scores:
grades = {"Alice": 85, "Bob": 92, "Charlie": 78}

Perform these tasks:
- Add a new student "Diana" with score 88
- Update Bob's score to 95
- Print all students with scores above 80
- Calculate and print the average score

---

`````python
# Name: Shada Jaleel Thundakkachi 
# MISIS: M01096649
grades = {"Alice": 85, "Bob": 92, "Charlie": 78}

grades["Diana"] = 88 #adding a new student named "diana" with 88 score.
grades["Bob"] = 95   #Updating bob's score to 95. 

print("updates grades:", grades) 

#to display the names of students who scored above 80. 
print("Students with scores above 80:") 
for student in grades: 
    if grades[student] >80: 
        print(student, "=", grades[student]) 
        
total= 0

for score in grades.values(): # add all scores 
    total+= score 
average= total / len(grades) 
print("Average:", average) 
`````

### Activity 12: Reversing Clothing Stock List

Create a program that reads the `clothing_stock.txt` file and stores all the clothing items in a list. Then reverse the order of the items and write the reversed list into a new file called `reversed_clothing_stock.txt`.

**Steps:**
1. Read all clothing items into a list.
2. Reverse the list.
3. Write the reversed list into a new file.
<hr>

`````python
# Name: Shada Jaleel Thundakkachi 
# MISIS: M01096649
with open("clothing_stock.txt", "r") as f:
#read all the lines and store them in a list called "items"
    items= f.readlines()   

items.reverse() #reverse of the items in the list. 

#open the file in write mode to save the reversed list. 
with open("reversed_clothing_stock.txt", "w") as f: 
    for item in items: 
        f.write(item) 

print("Reversed clothingitems have been written to reversed_clothing_stock.txt") 
`````
