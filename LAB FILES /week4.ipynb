3. Create a function taking one parameters for a password. The function would convert each character of the password into corresponding ASCII numbers and return one string composed of the ASCII numbers instead

````python
def ascii(password): 
    result= [] # empty string
    for i in password:
        result += str(ord(i)) #add the ascii number of the letter 
    return result

password= input("Enter password: ") 

print("ASCII password", "=", ascii(password))
````

4. Create a function (or set of functions ) taking one parameter for a password.

  - the password contains at least one lower case letter,
  - the password contains at least one digit and one special character like for example ‘@’.

````python
password= input("Enter password: ") 

def check_password(password): 
    if " "  in password: # checks whether there are any spaces 
        print("password should not contain any spaces") 
        return 
        
    for x in password: 
        if x.isupper(): 
            uppercase= True 
        elif x.islower(): 
            lowercase= True 
        elif x.isdigit(): 
            digit= True 
        elif x in "@#$%^&*!": 
            specialchar= True 

    print("Uppercase", "=", uppercase) 
    print("Lowercase", "=", lowercase) 
    print("Digit", "=", digit) 
    print("Special characters", "=", specialchar) 

check_password(password)
````

16. (Binary to hex) Write a function that parses a binary number and converts itinto a hex number. The function header is: def binaryToHex(binaryValue):
Write a test program that prompts the user to enter a binary number and displays the corresponding hexadecimal value.



````python
# Name: Shada Jaleel Thundakkachi 
# MISIS: M01096649

# convert binary string into decimal number using int() with base 2 
# Then convert decimal into hexadecimal using hex() 
def binaryToHex(binaryValue): 
    return hex(int(binaryValue, 2))[2:].upper()

binaryValue= input("Enter number: ") 

print("Hexadecimal value", "=", binaryToHex(binaryValue))  
````
